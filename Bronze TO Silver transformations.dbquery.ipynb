```sql
%sql
create or replace table workspace.silver.voltas AS
SELECT
    raw_payload:meeting_key::INT            AS chave_reuniao,
    raw_payload:session_key::INT            AS chave_sessao,
    raw_payload:driver_number::INT          AS numero_piloto,
    raw_payload:lap_number::INT             AS numero_volta,
    raw_payload:date_start::TIMESTAMP       AS data_inicio,
    raw_payload:duration_sector_1::DOUBLE   AS duracao_setor_1,
    raw_payload:duration_sector_2::DOUBLE   AS duracao_setor_2,
    raw_payload:duration_sector_3::DOUBLE   AS duracao_setor_3,
    raw_payload:lap_duration::DOUBLE        AS duracao_volta,
    raw_payload:is_pit_out_lap::BOOLEAN      AS volta_saida_pit,
    raw_payload:st_speed::DOUBLE            AS velocidade_armadilha,
    _driver_number                          AS _numero_piloto,
    _ingested_at                            AS _ingerido_em,
    current_timestamp()                     AS _processado_em
FROM workspace.bronze.laps_raw
WHERE raw_payload IS NOT NULL
  AND raw_payload:lap_number IS NOT NULL;
```
